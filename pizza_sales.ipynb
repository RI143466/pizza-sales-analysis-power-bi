```python
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
y=df['unit_price']
plt.plot(y)
plt.show()
```

```python
a=df.loc[200:300,'unit_price']
plt.plot(a,'m:o')
plt.show()
```

```python
cat=df['pizza_category'].value_counts()
x=cat.values
y=cat.index
a=['red','blue','pink']
plt.figure(figsize=(8,6),dpi=100)
plt.pie(x,labels=y,colors=a)
plt.legend(y)
plt.title('category')
plt.show()
```

```python
a=df[df['total_price']>30]['pizza_size']
b=a.value_counts()
x=b.index
y=b.values
plt.figure(figsize=(8,6),dpi=100)
plt.title("Pizza price by  size greater than 30")
plt.xlabel('Size')
plt.ylabel('Total_price')
plt.bar(x,y,color='r')
plt.show()
```

```python
# top 10 best selling pizzas
best_pizzas=df.groupby('pizza_name')['quantity'].sum().head(10).sort_values()
best_pizzas.plot(kind='bar')
plt.title("Top 10 pizzas")
plt.show()
```
